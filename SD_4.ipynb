```python
# Implementasi Queue menggunakan List
class QueueList:
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        return None
    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh penggunaan
queue = QueueList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Queue setelah enqueue :", queue.queue)
print("Dequeued element :", queue.dequeue())
print("Front element :", queue.front())
print("Rear element :", queue.rear())
print("Queue size :", queue.size())
```

```python
from collections import deque

#Implementasi Queue menggunakan deque
class QueueDeque:
    def __init__(self):
        self.queue = deque()

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        return self.queue.popleft() if not self.is_empty() else None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh penggunaan
queue = QueueDeque()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Queue setelah enqueue :", queue.queue)
print("Dequeued element :", queue.dequeue())
print("Front element :", queue.front())
print("Rear element :", queue.rear())
print("Queue size :", queue.size())
```

```python
class Node:
    def __init__ (self, data):
        self.data = data
        self.next = None

class QueueLinkedList:
    def __init__ (self):
        self.front = self.rear = None

    def enqueue (self, data):
        new_node = Node (data)
        if self.rear is None:
            self.front = self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node

    def dequeue (self):
        if self.front is None:
            return None
        dequeued_element = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return dequeued_element

    def is_empty(self):
        return self.front is None

    def front_element (self):
        return self.front.data if self.front else None

    def rear_element(self):
        return self.rear.data if self.rear else None
# Contoh penggunaan
queue = QueueLinkedList()
queue.enqueue (10)
queue.enqueue (20)
queue.enqueue (30)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front_element ())
print("Rear element:", queue.rear_element ())
```

```python
class TicketQueue:
    def __init__(self, max_size=10):
        self.queue = []
        self.max_size = max_size

    def enqueue(self, ticket_id):
        if len(self.queue) < self.max_size:
            self.queue.append(ticket_id)
            print(f"Pelanggan {ticket_id} telah ditambahkan ke antrian.")
        else:
            print("Antrian sudah penuh. Silakan tunggu hingga ada slot kosong.")

    def dequeue(self):
        if not self.is_empty():
            served_customer = self.queue.pop(0)
            print(f"Pelanggan {served_customer} telah dilayani.")
            return served_customer
        else:
            print("Antrian kosong. Tidak ada pelanggan yang dapat dilayani.")

    def front(self):
        if not self.is_empty():
            return self.queue[0]
        else:
            return None

    def rear(self):
        if not self.is_empty():
            return self.queue[-1]
        else:
            return None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def serve_vip(self, ticket_id):
        if '7' in ticket_id:
            print(f"Pelanggan VIP {ticket_id} dilayani tanpa antri.")
        else:
            print(f"Pelanggan {ticket_id} bukan VIP dan harus mengantri.")

# Contoh penggunaan
if __name__ == "__main__":
    ticket_queue = TicketQueue()

    # Menambahkan pelanggan
    ticket_queue.enqueue("Yo06")
    ticket_queue.enqueue("Yo07")
    ticket_queue.enqueue("Yo08")
    ticket_queue.enqueue("Yo09")
    ticket_queue.enqueue("Yo10")
    ticket_queue.enqueue("Yo11")
    ticket_queue.enqueue("Yo12")
    ticket_queue.enqueue("Yo13")
    ticket_queue.enqueue("Yo14")
    ticket_queue.enqueue("Yo15")

    # Mencoba menambahkan pelanggan saat antrian penuh
    ticket_queue.enqueue("Yo16")

    # Melayani pelanggan
    ticket_queue.dequeue()
    ticket_queue.dequeue()

    # Menampilkan pelanggan pertama dan terakhir
    print(f"Pelanggan pertama dalam antrian: {ticket_queue.front()}")
    print(f"Pelanggan terakhir dalam antrian: {ticket_queue.rear()}")

    # Mengecek apakah antrian kosong dan ukuran antrian
    print(f"Apakah antrian kosong? {ticket_queue.is_empty()}")
    print(f"Jumlah pelanggan dalam antrian: {ticket_queue.size()}")

    # Melayani pelanggan VIP
    ticket_queue.serve_vip("Yo07")
    ticket_queue.serve_vip("Yo08")
```

```python
class CopyQueue:
    def __init__(self):
        self.queue = []

    def enqueue(self, order_id, num_pages):
        self.queue.append((order_id, num_pages))
        print(f"Pesanan {order_id} dengan {num_pages} halaman telah ditambahkan ke antrian.")

    def dequeue(self):
        if not self.is_empty():
            order_id, num_pages = self.queue.pop(0)
            print(f"Pesanan {order_id} dengan {num_pages} halaman telah dilayani.")
            # Jika pesanan lebih dari 100 halaman, proses tahap kedua
            if num_pages > 100:
                remaining_pages = num_pages - 100
                print(f"Pesanan {order_id} memiliki sisa {remaining_pages} halaman, akan diproses kembali.")
                self.enqueue(order_id, remaining_pages)
            return order_id
        else:
            print("Antrian kosong. Tidak ada pesanan yang dapat dilayani.")

    def front(self):
        if not self.is_empty():
            return self.queue[0]
        else:
            return None

    def rear(self):
        if not self.is_empty():
            return self.queue[-1]
        else:
            return None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def fast_track(self, order_id, num_pages):
        if num_pages < 10:
            print(f"Pesanan {order_id} dengan {num_pages} halaman diproses tanpa antri.")
        else:
            self.enqueue(order_id, num_pages)

# Contoh penggunaan
if __name__ == "__main__":
    copy_queue = CopyQueue()

    # Menambahkan pesanan fotokopi
    copy_queue.fast_track("Yog06", 5)  # Fast track
    copy_queue.enqueue("Yog07", 50)
    copy_queue.enqueue("Yog08", 120)  # Pesanan lebih dari 100 halaman
    copy_queue.enqueue("Yog09", 15)

    # Melayani pesanan
    copy_queue.dequeue()  # Melayani Yog06
    copy_queue.dequeue()  # Melayani Yog07
    copy_queue.dequeue()  # Melayani Yog08 (akan memproses sisa halaman)

    # Menampilkan pesanan pertama dan terakhir
    print(f"Pesanan pertama dalam antrian: {copy_queue.front()}")
    print(f"Pesanan terakhir dalam antrian: {copy_queue.rear()}")

    # Mengecek apakah antrian kosong dan ukuran antrian
    print(f"Apakah antrian kosong? {copy_queue.is_empty()}")
    print(f"Jumlah pesanan dalam antrian: {copy_queue.size()}")
```

```python
class ParkingQueue:
    def __init__(self, max_capacity=15):
        self.queue = []
        self.max_capacity = max_capacity

    def enqueue(self, vehicle_number):
        if len(self.queue) < self.max_capacity:
            self.queue.append(vehicle_number)
            print(f"Kendaraan {vehicle_number} telah masuk ke parkiran.")
        else:
            print("Parkiran sudah penuh. Kendaraan baru harus menunggu hingga ada kendaraan yang keluar.")

    def dequeue(self):
        if not self.is_empty():
            vehicle_number = self.queue.pop(0)
            print(f"Kendaraan {vehicle_number} telah keluar dari parkiran.")
            return vehicle_number
        else:
            print("Parkiran kosong. Tidak ada kendaraan yang dapat keluar.")

    def front(self):
        if not self.is_empty():
            return self.queue[0]
        else:
            return None

    def rear(self):
        if not self.is_empty():
            return self.queue[-1]
        else:
            return None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def park_priority(self, vehicle_number):
        # Misalkan kendaraan ramah lingkungan ditandai dengan awalan "EL" (Electric)
        if vehicle_number.startswith("EL"):
            print(f"Kendaraan ramah lingkungan {vehicle_number} mendapatkan prioritas masuk parkiran.")
            self.enqueue(vehicle_number)
        else:
            self.enqueue(vehicle_number)

# Contoh penggunaan
if __name__ == "__main__":
    parking_queue = ParkingQueue()

    # Menambahkan kendaraan
    parking_queue.park_priority("Ma06")  # Kendaraan biasa
    parking_queue.park_priority("EL01")  # Kendaraan ramah lingkungan
    parking_queue.park_priority("Ma07")
    parking_queue.park_priority("Ma08")
    parking_queue.park_priority("Ma09")
    parking_queue.park_priority("Ma10")
    parking_queue.park_priority("Ma11")
    parking_queue.park_priority("Ma12")
    parking_queue.park_priority("Ma13")
    parking_queue.park_priority("Ma14")
    parking_queue.park_priority("Ma15")
    parking_queue.park_priority("Ma16")  # Kendaraan biasa
    parking_queue.park_priority("EL02")  # Kendaraan ramah lingkungan

    # Melayani kendaraan keluar
    parking_queue.dequeue()  # Mengeluarkan kendaraan pertama
    parking_queue.dequeue()  # Mengeluarkan kendaraan kedua

    # Menampilkan kendaraan pertama dan terakhir
    print(f"Kendaraan pertama dalam antrian: {parking_queue.front()}")
    print(f"Kendaraan terakhir dalam antrian: {parking_queue.rear()}")

    # Mengecek apakah parkiran kosong dan ukuran antrian
    print(f"Apakah parkiran kosong? {parking_queue.is_empty()}")
    print(f"Jumlah kendaraan dalam parkiran: {parking_queue.size()}")
```
